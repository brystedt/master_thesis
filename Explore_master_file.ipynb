```python
import pandas as pd
import numpy as np
```

```python
data_master_file = pd.read_csv('data_master_file_191014.csv', sep=';')
```

```python
data_master_file
```

```python
for col in data_master_file.columns: 
    print(col) 
```

```python
data_col_select = data_master_file[['patient', 'x_case', 'basename', 'dicom_studydate', 'x_diadate', 'dicom_imagelaterality', 'dicom_viewposition', 'x_cancer_laterality', 'dicom_imagerpixelspacing']]
data_cancers = data_col_select.loc[data_col_select['x_case'] == 1]
data_cancers
```

```python
data_rel = data_cancers[['patient','basename', 'dicom_imagelaterality', 'dicom_viewposition', 'dicom_imagerpixelspacing']]
data_rel['filename'] = data_rel['basename'].str.replace('.dcm','.png')
data_rel = data_rel.drop(columns=['basename'])
data_rel
```

```python
data_rel['d_'] = data_rel['dicom_imagerpixelspacing'].str.replace('[','').str.replace(']','').str.replace("'",'').str.split(",")

data_rel['d1'] = data_rel['d_'].apply(lambda x: float(x[0]))
data_rel['d2'] = data_rel['d_'].apply(lambda x: float(x[1]))
print(np.average(data_rel['d1']))
print(np.average(data_rel['d2']))
```

```python
data_rel.to_csv('master_cases',index=False)
```

```python
pd.set_option('display.max_colwidth', -1)
data_rel[['basename']].head(1)
```

```python
tumors = pd.read_csv('Tumors.txt', header=None)
tumors.columns = ['tumor_filename']
tumor_img = tumors[~tumors['tumor_filename'].str.contains('mask')]
tumor_img['key'] = tumor_img['tumor_filename'].str.replace('.png','')
tumor_img
```

```python
priors = pd.read_csv('Priors.txt', header=None)
priors.columns = ['prior_filename']
prior_img = priors[~priors['prior_filename'].str.contains('mask')]
prior_img['key'] = prior_img['prior_filename'].str.replace('.png','')
prior_img
```

```python
full_tumor_img = pd.merge(data_rel,tumor_img,how='outer',on='key').dropna()
full_tumor_img
```

```python
full_prior_img = pd.merge(data_rel,prior_img,how='outer',on='key').dropna()
full_prior_img
```

```python
full_prior_img['d_'] = full_prior_img['dicom_imagerpixelspacing'].str.replace('[','').str.replace(']','').str.replace("'",'').str.split(",")

full_prior_img['d1'] = full_prior_img['d_'].apply(lambda x: float(x[0]))
full_prior_img['d2'] = full_prior_img['d_'].apply(lambda x: float(x[1]))
print(np.average(full_prior_img['d1']))
print(np.average(full_prior_img['d2']))

```

```python
full = pd.merge(full_tumor_img,full_prior_img,how="outer",on=['patient','dicom_imagelaterality','dicom_viewposition'])#,'dicom_imagerpixelspacing'])
full = full.dropna()
full = full.drop(columns=['basename_x','basename_y','key_x','key_y'])
full
```
